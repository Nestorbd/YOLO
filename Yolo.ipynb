#### Predicción

```python
from ultralytics import YOLO

image_paths = ["test_img/cats and dogs.jpg", "test_img/cat and dog 1.jpg", "test_img/cat.jpg"]
# Cargar el modelo YOLO preentrenado
model = YOLO("results/weights/best.pt")
# Realizar la predicción en una imagen
results = model.track(source=image_paths, conf=0.3,
iou=0.5) 
i = 0
# Process results list
for result in results:
    i+=1
    boxes = result.boxes  # Boxes object for bounding box outputs
    result.show()  # display to screen
    result.save(filename= f"predicts/predict {i}.jpg")
    for box in boxes:
        print("Box:", box) 
```

### Predicción con webcam

```python
from ultralytics import YOLO

# Cargar el modelo YOLO preentrenado
model = YOLO("results/weights/best.pt")
# Realizar la predicción en una imagen
results = model.track(source=1, conf=0.3,
iou=0.5, stream=True, show=True) 

# Process results list
for result in results:
    boxes = result.boxes
    for box in boxes:
        print("Box:", box) 
```
